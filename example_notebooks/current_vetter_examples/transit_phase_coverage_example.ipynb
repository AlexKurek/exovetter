```python
import lightkurve as lk
from exovetter.tce import Tce
from exovetter import vetters as vet
import matplotlib.pyplot as plt
%matplotlib notebook
```

## Running the Transit Phase Coverage vetter on a known exoplanet

We use the previously downloaded lightcurve and TCE of wasp 18b to run the vetter (see the TCE and lightcurve example notebook)  

```python
lcf = lk.read('wasp18b_lc.fits')
lc = lcf.SAP_FLUX.remove_nans().remove_outliers()
flat = lc.flatten(window_length=81)

tce = Tce.from_json('wasp18b_tce')
```

```python
# View the tce and flattened lightcurve we're vetting: 
display(tce)

flat.plot()
plt.show()
```

## Transit Phase Coverage

Description

```python
# We initialize the transit phase coverage object 
# using the name of whatever array we want to be flux from the light_curve object 
transit_phase_coverage = vet.TransitPhaseCoverage(lc_name="flux")

# NOTE the TransitPhaseCoverage vetter also has access to the nbins and ndur parameters

# We then run the vetter on the desired TCE and lightcurve which results in a dictionary of relevent metrics
transit_phase_coverage_results = transit_phase_coverage.run(tce,flat)

print("Results from Transit Phase Coverage:")
display(transit_phase_coverage_results)
```

```python
# We can also use the built in plot method to visualize the results
transit_phase_coverage.plot()

# The plot parameter can also be set to True when running the vetter, 
# ie transit_phase_coverage.run(tce, flat, plot=True)
```
